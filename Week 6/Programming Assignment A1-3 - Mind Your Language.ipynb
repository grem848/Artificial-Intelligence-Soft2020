### Problem: 
Some text documents meant for distribution in Internet contain bad language and need
to be either corrected or deleted. The language can be controlled automatically by AI software,
which uses linguistic analysis techniques. As an example, a web service for searching inappropriate
words in a text is published at http://www.wdylike.appspot.com/

### Task: 
Create a Python program, which

• opens a text file from a local folder

• reads the text stored in the file

• connects to the remote service, which checks for profanity in the language

• sends the content of your file to this service

• gets and prints the returned results

```python
# Import request and parse
import urllib.request
import urllib.parse

# read a txt file function
def read_text(filename):
    # open text file from filename
    textfile = open(filename)
    
    # read file content
    file_content = textfile.read()
    
    # print file content
    print("File content to check: " + file_content)
    
    # close opened file
    textfile.close()
    
    # run check_profanity
    check_profanity(file_content)
    print("--------")

# check a text for profanities using wdylike.appspot.com
def check_profanity(text_to_check):
    # parse text content
    encoded_text = urllib.parse.quote(text_to_check, 'utf-8')
    
    # define profanity API url and add text as parameter
    APIurl = "http://www.wdylike.appspot.com/?q=" + encoded_text
    
    # open connection to the API
    connection = urllib.request.urlopen(APIurl)
    
    # read connection response (true or false)
    output = connection.read().decode('utf-8')
    
    # print response
    print("Contains profanity: " + output)
    
    # close connection
    connection.close()
    
    # check if response is true
    if "true" in output:
        print("Profanity Alert!")
    
    # check if response is false
    elif "false" in output:
        print("This document has no curse words!")
        
    # print if a problem exists with the response
    else:
        print("Could not scan the document properly.")


# Create, write and close a text file containing a profanity
f1 = open("DirtyText.txt","w")
f1.write("Fuck this text is dirty")
f1.close()

# Create, write and close a text file containing a clean text
f2 = open("CleanText.txt","w")
f2.write("This text is very clean")
f2.close()

# call defs
read_text(f1.name)
read_text(f2.name)
```
